```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("student_spending234443.csv")

# Average instead of count
weekend_avg = df.groupby('Weekend')['Average'].mean()

# Plot
plt.figure()
weekend_avg.plot(kind='bar')
plt.title('Average Expense on Weekday vs Weekend')
plt.xlabel('Weekend')
plt.ylabel('Average Expense')
plt.xticks(rotation=0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV file
df = pd.read_csv('student_spending234443.csv')

# Count values of category column (example: 'Category')
category_data = df['Total'].value_counts()

# Plot pie chart
plt.figure()
category_data.plot(kind='pie', autopct='%1.1f%%')

plt.title('Expense Categories')
plt.ylabel('')  # removes default ylabel

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load file
df = pd.read_csv('student_spending234443.csv')

# Select only expense columns
expense_cols = [
    'Housing', 'food', 'Junk food', 'Transport', 'Books',
    'Entertainment', 'Personal', 'Technology', 'Recharge',
    'Gym', 'Weekend', 'Health', 'Miscellaneous'
]

# Sum each category
expense_data = df[expense_cols].sum()

# Plot pie chart
plt.figure()
expense_data.plot(kind='pie', autopct='%1.1f%%')

plt.title('Overall Expense Distribution')
plt.ylabel('')

plt.show()
```
