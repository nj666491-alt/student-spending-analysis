```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("student_spending234443.csv")

# Average instead of count
weekend_avg = df.groupby('Weekend')['Average'].mean()

# Plot
plt.figure()
weekend_avg.plot(kind='bar')
plt.title('Average Expense on Weekday vs Weekend')
plt.xlabel('Weekend')
plt.ylabel('Average Expense')
plt.xticks(rotation=0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV file
df = pd.read_csv('student_spending234443.csv')

# Count values of category column (example: 'Category')
category_data = df['Total'].value_counts()

# Plot pie chart
plt.figure()
category_data.plot(kind='pie', autopct='%1.1f%%')

plt.title('Expense Categories')
plt.ylabel('')  # removes default ylabel

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load file
df = pd.read_csv('student_spending234443.csv')

# Fix column names (important)
df.columns = df.columns.str.strip().str.title()

# Select expense columns
expense_cols = [
    'Housing', 'Food', 'Junk Food', 'Transport', 'Books',
    'Entertainment', 'Personal', 'Technology', 'Recharge',
    'Gym', 'Weekend', 'Health', 'Miscellaneous'
]

# Sum each category
expense_data = df[expense_cols].sum()

# Sort values
expense_data = expense_data.sort_values(ascending=False)

# Take top 6
top_expenses = expense_data[:6]

# Add others
others = expense_data[6:].sum()
top_expenses['Others'] = others

# Plot
plt.figure(figsize=(7,7))

top_expenses.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90
)

plt.title('Overall Expense Distribution (Simplified)')
plt.ylabel('')

plt.show()
```
